```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Part a 
#from Ryden 7.11, q = O_r + (1/2)O_m - O_l
#for a benchmark Universe, q = 0 + (1/2)*.3 - .7 = -0.55 
#for a matter only Universe, q = 0 + (1/2)*1 - 0 = .5 
```

```python
#create luminosity distance equation 
c=2.998*10**5
Ho=70
def Lum_distance(q,z):
    DL=(c/Ho)*z*(1-((1+q)/2)*z)*(1+z)
    return DL
```

```python
redshifts=np.linspace(0,1,100)
```

```python
#using CosmoCalc http://www.astro.ucla.edu/~wright/CosmoCalc.html
#entering the appropriate values for O_m, Ho = 70 and clicking "flat":
z_cosmo=[.1,.2,.3,.4,.5,.6,.7,.8,.9]
dl_bm=[460.3,980.1,1552.7,2172.1,2832.8,3530.0,4259.4,5017.4,5800.9]
dl_matter=[438.5,895.6,1369.0,1856.8,2357.6,2870.1,3393.2,3925.9,4467.5]
```

```python
#plot all the things
plt.figure(figsize=(7,7))
plt.plot(redshifts,Lum_distance(-.55,redshifts),label='benchmark Universe',c='b')
plt.scatter(z_cosmo,dl_bm,c='b',marker='o')
plt.plot(redshifts,Lum_distance(.5,redshifts),label='matter only',c='g')
plt.scatter(z_cosmo,dl_matter,c='g')
plt.legend(loc=2)
plt.xlabel('redshift',fontsize=18)
plt.ylabel('Luminosity Distance (Mpc)',fontsize=18)
plt.xlim(0,1)
plt.ylim(0,7000)
```

```python
#we can see that the approximation is only valid to z~0.1 or so 
```

```python
#Part b - comparing Taylor approximation to linear approximation
```

```python
def Lum_distance_linear(z):
    dl=(c/Ho)*z
    return dl
```

```python
z_lin=np.linspace(0,.2,100)
```

```python
#plot all the things
plt.figure(figsize=(7,7))
plt.plot(z_lin,Lum_distance(-.55,z_lin),label='benchmark Universe',c='b')
plt.plot(z_lin,Lum_distance(.5,z_lin),label='matter only',c='g')
plt.plot(z_lin,Lum_distance_linear(z_lin),label='linear approximation')
plt.legend(loc=2)
plt.xlabel('redshift',fontsize=18)
plt.ylabel('Luminosity Distance (Mpc)',fontsize=18)
plt.xlim(0,.2)
```

```python
#Where do the equations differ by > 10%?
for red in redshifts: 
    if np.abs(Lum_distance_linear(red) - Lum_distance(-.55,red)) > 0.1*Lum_distance(-.55,red):
        print 'The benchmark formula deviates from the linear approximation by 10% at z = ', round(red,2)
        print 'The linear approximation distance at this redshift is', round(Lum_distance_linear(red),2), 'Mpc.'
        print 'The benchmark distance at this redshift is', round(Lum_distance(-.55,red),2), 'Mpc.'
        print
        break
for red in redshifts: 
    if np.abs(Lum_distance_linear(red) - Lum_distance(.5,red)) > 0.1*Lum_distance(.5,red):
        print 'The matter-only formula deviates from the linear approximation by 10% at z = ', round(red,2)
        print 'The linear approximation distance at this redshift is', round(Lum_distance_linear(red),2), 'Mpc.'
        print 'The matter-only distance at this redshift is', round(Lum_distance(.5,red),2), 'Mpc.'
        break
```

```python
#Part c

#Absolute magnitude is defined as the apparant magnitude of an object as viewed from a distance of 10 pc = 10^-5 Mpc
#Using Pogson's equation - m1 - m2 = -2.5log(F1/F2), and Since F1/F2 = (D2/D1)^2:

# m - M = -2.5log((10^-5/D)^2) = -5log(10^-5/D) = +5log(D/10^-5)
# m - M = 5[log(D) - log(10^-5)] = 5[log(D) - -5*log(10)] = 5[log(D) + 5] = 5*log(D) + 25
```

```python
# Part d 
#Simply using log rules log(A/B) = log(A) - log(B), and that D ~ cz/Ho:
#m - M = 5log(cz/Ho) + 25 = 5log(cz) - 5log(Ho) + 25

#To determine Ho from the above equation, we could plot m-M vs log(cz), fit an equation to the plot and calculate the 
#y-intercept value 'b', then b = -5log(Ho) + 25, or Ho = 10^(25-b)/5
```

```python
#Part e
#entering apparent and deltam values for the b-band magnitudes, removing outliers:

m=np.array([17.39,18.03,16.32,14.72,17.65,15.84,16.11,18.84,19.40,15.17])
delta_m=np.array([1.21,1.11,1.03,.94,1.35,1.88,1.17,1.18,1.32,.82])
logcz=np.array([4.079,4.227,3.971,3.621,4.124,3.463,3.879,4.351,4.485,3.781])
```

```python
#equation from Phillips 1993 for determining absolute magnitudes
def abs_mag(dmb):
    M = -21.726 + 2.698*dmb
    return M
```

```python
#fit polynomial to data - y = mx + b, where y = m-M, m = 5, x = log(cz), and b = -5log(Ho) + 25
slope,b=np.polyfit(logcz,m-abs_mag(delta_m),1)
Ho_calc = 10**((25-b)/5)
```

```python
cz=np.linspace(3.2,4.7,10)

f=plt.figure(figsize=(7,7))
plt.scatter(logcz,m-abs_mag(delta_m))
plt.plot(cz,slope*cz+b)
plt.xlabel('log(cz)',fontsize=18)
plt.ylabel('m-M (B-mag)',fontsize=18)
f.text(.2,.8,'H = %s'%round(Ho_calc,2),fontsize=16)
f.text(.2,.7,'m-M = %s log(cz) + %s'%(round(slope,2),round(b,2)),fontsize=16)
```
